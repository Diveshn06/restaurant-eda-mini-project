# **7. Study Correlation**

```python
num_columns = ['average_cost_for_two','price_range','aggregate_rating','votes','photo_count']
corr = df[num_columns].corr()
corr
```

```python
## Outcomes:

# average_cost_for_two <-> price_range (0.7859): it shows strong positive -> higher cost means higher price range
# average_cost_for_two <-> aggregate_rating(0.2323) : Weak positive -> expensive places slightly have a better rating
# average_cost_for_two <-> photo_count(0.3419) : Moderate positive -> expensive restaurants tend to have more photos.
# aggregate_rating <-> votes(0.2720) : Weak positive -> more votes usually means slightly better ratings
# aggregate_rating <-> price_range(0.2523) : Weak positive -> premium restaurants slightly tend to have better ratings.
# votes <-> photo_count(0.6550) : Strong positive -> more photos uploaded usually means more votes.
```

```python
# Creating Heatmap:

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between numerical columns")
plt.show()
```

# **9. Creating a new column 'region'**

```python
df['city'].unique()
```

```python
# Defining some cities :
city_region = {
    'Agra': 'North-East',
    'Delhi': 'North', 'New Delhi': 'North', 'Gurgaon': 'North', 'Noida': 'North',
    'Mumbai': 'West', 'Pune': 'West',
    'Bangalore': 'South', 'Chennai': 'South', 'Hyderabad': 'South',
    'Kolkata': 'East', 'Patna': 'East',
    'Bhopal': 'Central', 'Indore': 'Central'

}

# Apply mapping, using map
df['region'] = df['city'].map(city_region)

# Replace NaN with "Other"
df['region'] = df['region'].fillna('Other')

print(df[['city','region']].head(25))
```
